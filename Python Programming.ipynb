### 'Input' function

```python
name=input("What is your name?")
print (name)
```

```python
number1=input("Enter the first number")
number2=input("Enter the second number")
print (number1)
print (number2)
print (type(number1))
print (type(number2))
```

input fuction always take data as a string variable

Converting string input into a integer variable

```python
number1=int(input("Enter the first number"))
print (number1)
print (type(number1))
```

Converting string input into float variable

```python
coffee_price1=input("price of a coffee")
print (type(coffee_price1))
coffee_price2=float(input("price of a coffee"))
print (type(coffee_price2))
```
