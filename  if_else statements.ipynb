## Grading System that only accept 0-100 values

```python
marks=int(input("Marks"))
if marks>=0 and marks<=100:
    if marks>=90:
        print("A")
    elif marks>=80:
        print("B")
    elif marks>=70:
        print("C")
    elif marks>=60:
        print("D")
    elif marks>=0:
        print("F")
else:
    print("Invalid")
```
